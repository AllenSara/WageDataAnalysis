### Train Function

```python
from sklearn.metrics import fbeta_score, accuracy_score
```

```python
#כעת, נגדיר פונקציה שמטרתה להריץ אלגוריתם ספציפי על מדגם בגודל מסוים מתוך הנתונים.
#בונים את המודל ב train 80%
#בודקים את הדיוק שלו ב test 20%
#זאת הפונקציה שלנו:
def train_predict(learner, sample_size, X_train, y_train, X_test, y_test):
    #הגדרנו אובייקט שיביא לנו את התוצאות של התהליך כדי לקבל מדדים שונים של האלגוריתם.
    results = {}
    
    # Learner-זהו סוג האלגוריתם שאנחנו נריץ בפונקציה 
    #sample_size-זהו גודל הדגימה שלנו
    # X_train, y_train, X_test, y_test-השאר זה חלוקת הדאטה ל80% ו 20%
    
    #b - time the fit process, מדידת זמן של תהליך האימון של האלגוריתם
    #התחלת זמן המדידה
    start = time()
    #הגדרת משתנה למידה והכנסה לתוכו המודל של הלמידת מכונה שלנו
    learner = learner.fit(X_train[:sample_size], y_train[:sample_size])
    #סיום זמן הלמידה
    end = time()
    #הכנסת התוצאה לתוך האובייקט שיצרנו למעלה
    results['train_time'] = end - start
    # התהליך של המשין לרניג 
    
    #c - time the predict process, מדידת זמן של תהליך החיזוי של האלגוריתם 
    #התחלת זמן המדידה
    start = time()
    #הגדרת משתנה החיזוי של הטסט והכנסה לתוכו את המודל של הלמידת חיזוי שלנו לפי הנתונים שיש בX_TEST
    predictions_test = learner.predict(X_test)
    #הגדרת משתנה האימון של הטסט והכנסה לתוכו את המודל של הלמידת חיזוי שלנו לפי הנתונים שיש ב300 שורות של אימון כדי לבדוק אח"כ עד כמה זה מדויק
    predictions_train = learner.predict(X_train[:300])
    #סיום זמן הלמידה
    end = time()
    #הכנסת התוצאה לתוך האובייקט שיצרנו למעלה
    results['pred_time'] = end - start
    #התוצאות של המשין לרנינג 
    # כלומר הוא מוציא פה את החיזוי שלו (0 או 1) למשכורת מעל חמישים
    
    #d - save the results of the accuracy process, כאן אנחנו נותנים ניקוד עד כמה התוצאה שהוא קיבל באמת חזתה את התוצאה האמיתית
    #כאן הניקוד מתבצע על 300 השורות של הY_TRAIN
    results['acc_train'] = accuracy_score(y_train[:300], predictions_train)
    #וכאן הניקוד מתבצע על הY_TEST
    results['acc_test'] = accuracy_score(y_test, predictions_test)
    # הקטע ב2 השורות האלה שהוא משווה את התוצאה שלו עם התוצאה האמיתית ובסוף אומר עם הוא הצליח באמת
    
    #e - save the results of the FBetaScore process, שמירת הביצועים לפי שיטת FBETASCORE
    #אנחנו עושים כמו 2 השורות מקודם רק בשיטה החדשה שלנו
    results['f_train'] = fbeta_score(y_train[:300], predictions_train, beta=0.5)
    results['f_test'] = fbeta_score(y_test, predictions_test, beta=0.5)
       
    #f - print, הדפסת התוצאות של מה שהאלגוריתם נתן
    print ("{} trained on {} samples.".format(learner.__class__.__name__, sample_size))
        
    #g - results, החזרת התוצאות של הפונקציה
    return results
```

### Model Evaluation

```python
#import algorithm libraries, יבוא ספריות שהם המודלים לחיזוי שלנו
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import SGDClassifier

#initialize the three models, הבאנו 3 מודלים לחיזוי
clf_A = AdaBoostClassifier(random_state = 0)
clf_B = GradientBoostingClassifier(random_state = 0)
clf_C = SGDClassifier(random_state = 0)

#set the sizes, נתנו משתנים עם דגימות שונות.
#כאן נתנו 100% מהדאטה
samples_100 = int(len(y_train))
#כאן נתנו 10% מהדאטה
samples_10 = int(len(y_train) * 0.1)
#כאן נתנו 1% מהדאטה
samples_1 = int(len(y_train) * 0.01)


#results, עשינו לולאה שאיתה הכנסנו שהוא ירוץ על כל המודלים אחד אחרי השני.
#בנוסף עשינו שבאמצעות הפונקציה שלנו נקבל תוצאות של חיזוי עם אלגוריתם הלמידה שלנו
#כאן פתחנו אובייקט לקבלת התוצאות
results = {}
#לולאת FOR שאיתה אנחנו מריצים את המודלים השונים
for clf in [clf_A, clf_B, clf_C]:
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    for i, samples in enumerate([samples_1, samples_10, samples_100]):
        #כאן אנחנו משתמשים בפונקציה שהכנו למעלה כדי להפעיל את המודלים שייבאנו בשורה הראשונה ולקבל את תוצאות החיזוי של המודלים הללו
        results[clf_name][i] = train_predict(clf, samples, X_train, y_train, X_test, y_test)

#print graphs, הצגה של הנתונים מתוך האובייקט RESULTS
vs.evaluate(results)
```

We would use the SGD Classifier to get the fastest results. As is seen in the first table above, Model Training, the SGD Classifier model never took longer than a second to run, while the other models always took far longer to process the data. 
If time were not a factor, we would choose the Gradient Boosting Classifier, as it had the highest Accuracy and FBetaScores across all the data, which means it will most likely give us the most accurate results.

### Gaussian Mixture

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score #this measures how similar or close an object is to its own clusture in comparison to other clusters
```

```python
#לצורך ההבנה, קלאסטר זה בעצם קבוצה של מספר נתונים שנמצאים קרובים אחד לשני.
#בכך שאנחנו נותנים להם סיווג קבוצתי מסוים.

#כאן אנחנו בנינו מערך כדי לבדוק איזו אופציה לחלוקה לקלאסטרים היא הכי טובה. לכן יש לנו כאן כמה מספרים שאיתם נבדוק זאת
clusterOptions = [2,3,4,5,6,7]
#המשתנה הזה שומר את המרחק הכי גדול בניהם
max_score = 0
#לולאה שעוברת על הקלאסטרים בהתאם למיקום במערך שעשינו למעלה, וככה היא מנסה למצוא איזה קלאסטר הוא הכי טוב לחלוקה מדויקת של המשתנים שלנו
for i in clusterOptions :
    #כאן אנחנו מחלקים לפי 10000 שורות שבטבלה
    clusterer=GaussianMixture(n_components=i).fit (features_final[:10000])
    #השורה הזאת בודקת את המרחק בין הקלאסטרים השונים
    predict = clusterer.predict (features_final[:10000]) 
    #השורה הזאת נותנת ציון עד כמה החלוקה הזאת טובה
    score = silhouette_score(features_final[:10000],predict,metric='euclidean')
    #אם הציון יותר טוב מהציון הקודם, אז הוא ישמור את קלאסטר הזה ואת הציון ויעבור לקלאסטר הבא ואם לא אז הוא רק עובר לקלאסטר הבא
    if (score > max_score):
        max_score = score
        cluster = i
#בסוף אנחנו מדפיסים את קלאסטר היה הכי טוב ומה הציון שהוא קיבל
print(f"We tested all the clusters between 2 and 7. The cluster with the greatest silhoutte score is cluster number: {cluster} The silhoutte score is: {max_score}")
```
